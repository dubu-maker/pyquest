```python
# raise로 예외 발생시키기
def divide(a, b):
    if b == 0:
        raise ValueError("0으로 나눌 수 없습니다.")
    return a / b

try:
    result = divide(10, 0)
except ValueError as e:
    print(e)
```

```python
# 예외 클래스 상속받아 사용자 정의 예외 만들기
class NegativeNumberError(Exception):
    """음수를 입력했을 때 발생하는 사용자 정의 예외"""
    pass

def square_root(x):
    if x < 0:
        raise NegativeNumberError("음수의 제곱근을 계산할 수 없습니다.")
    return x ** 0.5

try:
    result = square_root(-9)
except NegativeNumberError as e:
    print(e)
```

```python
# French Flashcard App 완성
from tkinter import *
import pandas
import random

BACKGROUND_COLOR = "#B1DDC6"
current_card = {}
to_learn = {}

try:
    data = pandas.read_csv("flash-card-project-start/data/words_to_learn.csv")
except FileNotFoundError:
    original_data = pandas.read_csv("flash-card-project-start/data/french_words.csv")
    to_learn = original_data.to_dict(orient="records")
else:
    to_learn = data.to_dict(orient="records")

def next_card():
    global current_card, flip_timer
    window.after_cancel(flip_timer)
    current_card = random.choice(to_learn)
    canvas.itemconfig(card_title, text="French", fill="black")
    canvas.itemconfig(card_word, text=current_card["French"], fill="black")
    canvas.itemconfig(card_background, image=card_front_img)
    flip_timer = window.after(3000, func=flip_card)

def flip_card():
    canvas.itemconfig(card_title, text="English", fill="white")
    canvas.itemconfig(card_word, text=current_card["English"], fill="white")
    canvas.itemconfig(card_background, image=card_back_img)
    # 3초 후에 새로운 카드를 보여주기 위해 next_card를 호출
    window.after(3000, next_card)
    
def is_known():
    to_learn.remove(current_card)
    data = pandas.DataFrame(to_learn)
    data.to_csv("flash-card-project-start/data/words_to_learn.csv", index=False) # 인덱스 번호 자동 생성을 막는다.
    next_card()

window = Tk()
window.title("flashcard")
window.config(padx=50, pady=50, bg=BACKGROUND_COLOR)

flip_timer = window.after(3000, func=flip_card)

canvas = Canvas(width=800, height=526, bg=BACKGROUND_COLOR, highlightthickness=0)
card_front_img = PhotoImage(file="flash-card-project-start/images/card_front.png")
card_back_img = PhotoImage(file="flash-card-project-start/images/card_back.png")
card_background = canvas.create_image(400, 263, image=card_front_img)

card_title = canvas.create_text(400, 150, text="Title", font=("Ariel", 40, "italic"))
card_word = canvas.create_text(400, 263, text="Word", font=("Ariel", 60, "bold"))

canvas.grid(row=0, column=0, columnspan=2)

cross_image = PhotoImage(file="flash-card-project-start/images/wrong.png")
unknown_button = Button(image=cross_image, highlightthickness=0, command=next_card)
unknown_button.grid(row=1, column=0)

check_image = PhotoImage(file="flash-card-project-start/images/right.png")
known_button = Button(image=check_image, highlightthickness=0, command=is_known)
known_button.grid(row=1, column=1)

next_card()

window.mainloop()
```

```python
import requests
import random

def fetch_questions(category, limit, difficulty):
    url = f"https://the-trivia-api.com/api/questions?categories={category}&limit={limit}&difficulty={difficulty}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to retrieve data: {response.status_code}")
        return []

def quiz_game():
    category = input("Enter the category (e.g., science, history): ").strip().lower()
    limit = int(input("Enter the number of questions: "))
    difficulty = input("Enter the difficulty (easy, medium, hard): ").strip().lower()
    player_name = input("Enter your name: ").strip()

    questions = fetch_questions(category, limit, difficulty)

    for index, q in enumerate(questions):
        print(f"Question {index + 1}: {q['question']}")
        options = q['incorrectAnswers'] + [q['correctAnswer']]
        random.shuffle(options)

        for i, option in enumerate(options):
            print(f"{i + 1}. {option}")

        while True:
            try:
                answer = int(input("Your answer (choose the number): "))
                if 1 <= answer <= len(options):
                    break
                else:
                    print(f"Please enter a number between 1 and {len(options)}.")
            except ValueError:
                print("Invalid input. Please enter a number.")

        correct_answer = q['correctAnswer']
        if options[answer - 1] == correct_answer:
            print("Correct!\n")
        else:
            print(f"Wrong! The correct answer was: {correct_answer}\n")

quiz_game()
```

```python
# wx 모듈로 만든 계산기
import wx

class CalculatorFrame(wx.Frame):
    def __init__(self, *args, **kw):
        super(CalculatorFrame, self).__init__(*args, **kw)
        
        self.SetTitle("Simple Calculator")
        self.SetSize((400, 500))  # 창 크기 확대
        
        # 패널 생성
        panel = wx.Panel(self)

        # 수직 박스 레이아웃 (VBox)
        vbox = wx.BoxSizer(wx.VERTICAL)

        # 결과를 표시할 텍스트 컨트롤 생성
        self.text_ctrl = wx.TextCtrl(panel, style=wx.TE_RIGHT, size=(380, 80))  # 입력창 크기 확대
        font = wx.Font(36, wx.DEFAULT, wx.NORMAL, wx.BOLD)
        self.text_ctrl.SetFont(font)
        vbox.Add(self.text_ctrl, 0, wx.EXPAND | wx.ALL, 10)

        # GridSizer 생성 (5x4 그리드, 셀 간 수직/수평 간격 10 픽셀)
        grid_sizer = wx.GridSizer(5, 4, 10, 10)

        # 버튼 추가
        buttons = [
            ('', None), ('', None), ('', None), ('C', self.on_clear),
            ('7', self.on_number), ('8', self.on_number), ('9', self.on_number), ('/', self.on_operator),
            ('4', self.on_number), ('5', self.on_number), ('6', self.on_number), ('*', self.on_operator),
            ('1', self.on_number), ('2', self.on_number), ('3', self.on_number), ('-', self.on_operator),
            ('0', self.on_number), ('%', self.on_operator), ('=', self.on_equal), ('+', self.on_operator)
        ]

        operator_buttons = {'/', '*', '-', '+', '=', '%', 'C'}

        for label, handler in buttons:
            if label:  # 빈 셀에 대해서는 버튼을 만들지 않음
                button = wx.Button(panel, label=label)
                if handler:
                    button.Bind(wx.EVT_BUTTON, handler)

                # 연산자와 = 버튼에 색상을 적용
                if label in operator_buttons:
                    button.SetBackgroundColour(wx.Colour(220, 220, 220))  # 연한 회색
                    button.SetForegroundColour(wx.Colour(0, 0, 0))       # 검정색 텍스트

                grid_sizer.Add(button, 0, wx.EXPAND)
            else:
                grid_sizer.Add(wx.StaticText(panel), 0, wx.EXPAND)  # 빈 셀을 위한 투명한 위젯

        vbox.Add(grid_sizer, 1, wx.EXPAND | wx.ALL, 10)

        panel.SetSizer(vbox)

        self.current_operator = None
        self.first_operand = None

    def on_number(self, event):
        label = event.GetEventObject().GetLabel()
        current_value = self.text_ctrl.GetValue()
        self.text_ctrl.SetValue(current_value + label)

    def on_operator(self, event):
        label = event.GetEventObject().GetLabel()
        current_value = self.text_ctrl.GetValue()
        if not current_value.endswith(('+', '-', '*', '/', '%')):
            self.current_operator = label
            self.first_operand = float(current_value)
            self.text_ctrl.SetValue(current_value + label)

    def on_clear(self, event):
        self.text_ctrl.SetValue("")
        self.current_operator = None
        self.first_operand = None

    def on_equal(self, event):
        current_value = self.text_ctrl.GetValue()
        try:
            if self.current_operator:
                second_operand = float(current_value.split(self.current_operator)[-1])

                if self.current_operator == '+':
                    result = self.first_operand + second_operand
                elif self.current_operator == '-':
                    result = self.first_operand - second_operand
                elif self.current_operator == '*':
                    result = self.first_operand * second_operand
                elif self.current_operator == '/':
                    if second_operand != 0:
                        result = self.first_operand / second_operand
                    else:
                        result = "Error"
                elif self.current_operator == '%':
                    result = (self.first_operand * second_operand) / 100
                else:
                    result = "Error"

                self.text_ctrl.SetValue(str(result))
        except ValueError:
            self.text_ctrl.SetValue("Error")
        finally:
            self.current_operator = None
            self.first_operand = None

# 애플리케이션 실행
if __name__ == "__main__":
    app = wx.App(False)
    frame = CalculatorFrame(None, wx.ID_ANY)
    frame.Show(True)
    app.MainLoop()
```

```python
# 정답일시 녹색 배경이 오답일 시 빨간색 배경
from tkinter import *
import requests
import random
import html
import os  # Import the 'os' module
from openai import OpenAI

# OpenAI 클라이언트 인스턴스 생성
client = OpenAI(
    api_key=os.environ.get("OPENAI_API_KEY"),
)

BACKGROUND_COLOR = "#B1DDC6"
CORRECT_COLOR = "#4CAF50"  # 정답일 때의 배경색 (녹색)
INCORRECT_COLOR = "#F44336"  # 오답일 때의 배경색 (빨간색)
BUTTON_BACKGROUND_COLOR = "#9C27B0"  # 보기 버튼의 보라색 배경
BUTTON_ACTIVE_BACKGROUND_COLOR = "#7B1FA2"  # 버튼 활성화 시 배경색
BUTTON_TEXT_COLOR = "white"  # 버튼 텍스트 색상
current_question = {}
trivia_data = []
asked_questions = set()  # 이미 출제된 질문을 저장할 집합 / 중복을 제거한다.
api_source = ""  # 현재 사용 중인 API 출처를 저장할 변수

# 번역을 제외할 고유 명사 리스트
EXCLUDE_LIST = ["Pink Floyd", "Beatles", "Google", "Microsoft"]  # 필요에 따라 추가

# 제외할 카테고리 리스트
EXCLUDE_CATEGORIES = ["Entertainment: Video Games", "Video Games"]  # 게임 카테고리

def get_trivia_data():
    global trivia_data, api_source
    try:
        # Open Trivia API로 질문을 요청
        response = requests.get("https://opentdb.com/api.php?amount=10&type=multiple")
        response.raise_for_status()  # 응답 코드가 200이 아니면 예외 발생
        data = response.json()

        # Open Trivia API가 질문을 반환했는지 확인
        if data['response_code'] == 0: # 1: 오류, 잘못된 요청 2: 요청한 범주에 질문이 없음 3:사용할 수 없는 질문 유형 4: 세션 토큰과 관련된 오류 / http 응답 코드와 다르다
            trivia_data = data['results'] # results: 질문들을 담고 있는 리스트
            api_source = "Open Trivia API"  # 출처 설정
        else:
            raise ValueError("Open Trivia API returned no results.")
    
    except (requests.exceptions.RequestException, ValueError) as e:
        print(f"Open Trivia API failed: {e}")
        print("Switching to The Trivia API...")

        # The Trivia API로 대체
        response = requests.get("https://the-trivia-api.com/api/questions?limit=10&difficulty=medium")
        response.raise_for_status()
        trivia_data = response.json()
        api_source = "The Trivia API"  # 출처 설정

    # 카테고리 필터링
    trivia_data = [q for q in trivia_data if q.get('category', '') not in EXCLUDE_CATEGORIES]

def translate_to_korean(text):
    # 번역을 제외할 고유 명사가 포함된 경우, 해당 명사를 번역에서 제외합니다.
    for word in EXCLUDE_LIST:
        text = re.sub(r'\b' + re.escape(word) + r'\b', f'<no_translate>{word}</no_translate>', text)

    # 숫자나 단순 텍스트의 경우 번역을 건너뜁니다.
    if text.isdigit() or all(char.isdigit() or char.isspace() or char in ['(', ')', '-', '.', '%'] for char in text):
        return text

    # OpenAI API를 사용하여 텍스트를 번역
    chat_completion = client.chat.completions.create(
        messages=[
            {
                "role": "user",
                "content": f"Translate the following text to Korean: {text}",
            }
        ],
        model="gpt-4o",
    )
    translated_text = chat_completion.choices[0].message.content.strip()

    # 번역 제외 처리된 명사를 원래대로 복구
    translated_text = re.sub(r'<no_translate>(.*?)</no_translate>', r'\1', translated_text)

    return translated_text

def next_card():
    global current_question
    if len(trivia_data) == 0:
        get_trivia_data()
    
    current_question = trivia_data.pop(0) # .pop(0) : 리스트의 첫 번째 요소를 반환하고 삭제

    # 중복 문제 확인 및 방지
    question_text = html.unescape(current_question["question"])
    while question_text in asked_questions:
        if len(trivia_data) == 0:
            get_trivia_data()
        current_question = trivia_data.pop(0)
        question_text = html.unescape(current_question["question"])

    # 출제된 질문을 기록
    asked_questions.add(question_text)

    # 출처 출력
    print(f"Question source: {api_source} | Category: {current_question.get('category', 'Unknown')}")
    
    # 질문을 한국어로 번역
    translated_question = translate_to_korean(question_text)
    correct_answer = current_question["correct_answer"] if 'correct_answer' in current_question else current_question["correctAnswer"]
    incorrect_answers = current_question["incorrect_answers"] if 'incorrect_answers' in current_question else current_question["incorrectAnswers"]

    # 정답과 오답을 합쳐서 보기로 사용
    all_answers = incorrect_answers + [correct_answer]
    random.shuffle(all_answers)  # 보기를 랜덤으로 섞기

    # 각 보기를 한국어로 번역
    translated_answers = [translate_to_korean(answer) for answer in all_answers]

    canvas.itemconfig(card_word, text=translated_question)

    # 각 보기 버튼에 텍스트 설정
    for i in range(4):
        answer_buttons[i].config(text=translated_answers[i], command=lambda ans=all_answers[i]: check_answer(ans, correct_answer))

    # 배경색 초기화
    window.config(bg=BACKGROUND_COLOR)

def check_answer(selected_answer, correct_answer):
    if selected_answer == correct_answer:
        print("Correct!")
        window.config(bg=CORRECT_COLOR)
    else:
        print(f"Incorrect! The correct answer was: {correct_answer}")
        window.config(bg=INCORRECT_COLOR)
    
    window.after(1000, next_card)  # 1초 후에 다음 카드로 넘어가기

window = Tk()
window.title("flashcard")
window.config(padx=50, pady=50, bg=BACKGROUND_COLOR)

canvas = Canvas(width=800, height=526, bg=BACKGROUND_COLOR, highlightthickness=0)
card_word = canvas.create_text(400, 263, text="Word", font=("Ariel", 20, "bold"), width=680)

canvas.grid(row=0, column=0, columnspan=2)

# 객관식 보기를 위한 버튼들 생성
answer_buttons = []
for i in range(4):
    button = Button(window, text=f"Answer {i+1}", font=("Ariel", 12, "bold"), width=60, bg=BUTTON_BACKGROUND_COLOR, 
                    activebackground=BUTTON_ACTIVE_BACKGROUND_COLOR, fg=BUTTON_TEXT_COLOR, relief="raised", bd=5, cursor="hand2")
    button.grid(row=i+2, column=0, columnspan=2, pady=5)
    answer_buttons.append(button)

get_trivia_data()
next_card()

window.mainloop()
```
