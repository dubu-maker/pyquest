```python
# 로지스틱 회귀: 데이터가 2개의 범주 (고양이/고양이이 아님) 중 하나에 속할 확률을 예측한다.
# 시그모이드 함수를 적용한다.
# 입력값을 0과 1사이의 값으로 변환, 출력값은 확률로 해석할 수 있다.

# m은 훈련데이터의 샘플 수
# n은 특성 수(한 샘플이 가지는 특성 수)
```

# y_hat은 모델이 예측한 값으로, 0과 1사이의 확률로 표현된다. 입력 x와 모델의 파라미터 w, b를 사용하여 계산되며, 시그모이드 함수에 의해 출력된다
# z = wT + b (w는 가중치, b는 편향값)
w는 입력데이터의 차원에 맞는 nx 차원벡터
b는 편향값으로 실수

'시그모이드 함수'
z가 매우크면 값이 1에 가까워지고, 매우작으면 0에 가까워진다. 
# 출력값을 확률로 변환한다.

```python
'로지스틱 회귀 모델은 입력 데이터를 기반으로 이진 분류(0 또는 1)를 수행하는 모델이다.'
# 입력값 x와 가중치 w, 편향 b를 사용하여 선형 결합(x = wTx + b)을 계산한다.
# 그러나 출력값이 확률로 나와야 하므로 시그모이드 함수를 통해 반환된다

'시그모이드 함수는 로지스틱 회긔 모델의 예측값을 0과 1사이의 값으로 변환한다.'
# 선형 결합 z가 시그모이드 함수에 들어가서 예측값 y_hat(확률값)이 된다

# 손실 함수
'모델이 예측한 값 y_hat와 실제 값 y간의 차이의 측정하는 함수가 손실함수 이다.'

# 비용 함수
'비용함수는 전체 데이터셋에 대한 평균 손실을 나타낸다.'
'손실함수가 한 데이터 포인트에서의 차이를 측정한다면, 비용 함수는 모든 데이터 포인트에 대한 손실을 평균내어 모델이 전체 성능을 평가하는데 사용된다.'
# 모델을 훈련할 때는 이 비용 함수의 값을 최소화 하는 방향으로 학습한다.
```

# 로지스틱 회귀 모델은 입력 데이터 x를 받아 선형 결합 z = wTx + b를 계산한다
# 시그모이드 함수는 이 z값을 확률 값 y_hat로 변환한다
# 손실 함수는 예측된 값 y_hat과 실제 값 y사이의 차이를 측정한다
# 비용 함수는 손실 함수를 전체 데이터셋에 대해 평균을 내어, 모델의 전체 성능을 평가한다.

```python
# 기울기 하강 알고리즘로 학습세트의 매개변수 w를 학습한다. (경사 하강법)
'경사 하강법은 w와 b를 업데이트하면서, 비용 함수 J의 값이 최저점에 도달할 때까지 점진적으로 이동한다'
```

```python
# 도함수는 함수의 변화율을 나타내는 개념, 어떤 함수가 얼마나 빨리 변화하는지를 보여주는 함수
'도함수는 함수의 기울기 또는 변화율을 나타낸다. 어떤 함수가 어떤 점에서 얼마나 빠르게 증가하거나 감소하는지를 나타낸다'

# 비용함수의 도함수를 통해 기울기를 계산하고 그 기울기를 사용해 가중치 w와 편향 b를 업데이트하며 비용 함수의 값을 최소화하려고 한다
# 도함수가 양수면 함수는 증가하고 있는 것이므로, 기울기를 반대로 하여 내려간다
# 도함수가 음수면 함수는 감소하고 있는 것이므로, 기울기를 따라 내려간다.
```

# 경사하강법은 비용함수의 기울기(도함수)가 0이 되는 지점을 찾는다. 그 지점이 최저점

```python
import numpy as np
import matplotlib.pyplot as plt

# Function f(x) = x^2 and its derivative f'(x) = 2x
x = np.linspace(-10, 10, 400)
f_x = x**2
f_prime_x = 2*x

# Plotting the graph
plt.figure(figsize=(10, 6))

# Original function f(x) = x^2
plt.plot(x, f_x, label="f(x) = x^2", color='blue')

# Derivative f'(x) = 2x
plt.plot(x, f_prime_x, label="f'(x) = 2x (derivative)", color='red')

# Highlight the x and y axis
plt.axvline(0, color='black', linewidth=1)
plt.axhline(0, color='black', linewidth=1)

# Additional description
plt.title("Function f(x) = x^2 and its derivative f'(x) = 2x")
plt.xlabel("x values")
plt.ylabel("y values")
plt.legend()
plt.grid(True)

# Show the graph
plt.show()
```

```python
# 상수 함수는 변화가 없기 때문에, 미분하면 항상 0이 된다.

# 선형 함수의 미분: 선형 함수 f(x) = ax에서 미분하면, 기울기 a만 남는다.
'f(x) = 3x / 미분하면 기울기 3만 남는다.'

# 거듭제곱 함수의 미분(멱함수)
'f(x) = x^n에서, 미분하면 지수 n을 앞으로 내리고 지수를 1줄인다'
'f(x) = x^3 / 미분하면 3x^2'

# 합의 미분
'두 함수 f(x)와 g(x)의 합을 미분할때, 각각의 함수에 대해 따로 미분한 후 더해줍니다'
"f(x) = x^2 + 3x / f'(x) = 2x + 3'"

# 곱의 미분
'두 함수 f(x)와 g(x)의 곱을 미분할때, 첫번째 함수에 대한 미분과 두번째 함수에 대한 미분을 곱한 후 더한다'
"f(x) = x^2, g(x) = sin(x) / 2x * sin(x) + x^2 * cos(x)"
```

# 사인 함수를 미분하면 코사인이 남는다 (사인 함수의 기울기가 코사인 함수와 동일하다)
# 코사인 함수의 미분 결과는 마이너스 사인 함수, 기울기가 반대 방향

```python
import numpy as np
import matplotlib.pyplot as plt

# x 값 설정 (0부터 2pi까지)
x = np.linspace(0, 2 * np.pi, 400)
y_sin = np.sin(x)  # 사인 함수
y_cos = np.cos(x)  # 코사인 함수

# 미분한 함수
dy_sin = np.cos(x)  # 사인의 미분은 코사인
dy_cos = -np.sin(x)  # 코사인의 미분은 -사인

print(np.array_equal(y_sin, -dy_cos))  # 사인 함수와 코사인 함수의 미분은 서로 반대이다
print(np.array_equal(y_cos, dy_sin))  # 코사인 함수와 사인 함수의 미분은 서로 반대이다
```
