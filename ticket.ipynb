```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.edge.options import Options

# Edge 드라이버 경로 설정
edge_driver_path = './msedgedriver.exe'

# Edge 옵션 설정
edge_options = Options()

# 자동화 감지 우회 옵션 추가 (선택 사항)
edge_options.add_experimental_option('excludeSwitches', ['enable-automation'])
edge_options.add_experimental_option('useAutomationExtension', False)

# Service 객체 생성
service = Service(executable_path=edge_driver_path)

# Edge 브라우저 실행 (자동화 감지 우회)
driver = webdriver.Edge(service=service, options=edge_options)

# 이 시점에서 브라우저가 열리며, 수동으로 티켓팅 페이지로 이동할 수 있습니다.
print("Edge WebDriver가 실행되었습니다. 수동으로 페이지에 접속하세요.")

# 셀레늄 드라이버 열기
```

```python
from selenium.webdriver.common.by import By
import time

# Edge 옵션 설정
options = Options()
options.add_argument("user-agent=Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36")

def check_seat():
    while True:
        try:
            # 좌석이 나타나면 예매 가능한 좌석 선택 (s13 클래스가 예매 가능 좌석임을 가정)
            available_seat = driver.find_element(By.CSS_SELECTOR, 'div[class="s13"]')  # 클래스 확인 필요

            if available_seat:
                # 좌석 클릭
                available_seat.click()
                print("좌석을 선택했습니다.")

                # 좌석 선택 완료 버튼 클릭
                booking_button = driver.find_element(By.CLASS_NAME, 'booking')
                booking_button.click()  # 버튼 클릭
                break
        except:
            # 좌석이 없으면 페이지 새로고침
            print("빈 좌석이 없습니다. 새로고침 중...")
            time.sleep(1)
            driver.refresh()

# 좌석 확인 함수 실행
check_seat()

# 셀레늄으로 좌석 확인
```
