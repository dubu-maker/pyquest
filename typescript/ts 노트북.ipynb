네, 맞습니다! TypeScript 파일(`.ts`)을 JavaScript 파일(`.js`)로 컴파일한 후, 그 컴파일된 JavaScript 파일을 HTML 파일에서 로드해서 웹 페이지에서 실행합니다. 그래서 이 과정에는 **세 개의 파일**이 필요합니다:

1. **`index.html`**: 브라우저에서 보여줄 HTML 페이지
2. **`app.ts`**: TypeScript 코드가 작성된 파일
3. **`app.js`**: TypeScript에서 컴파일된 JavaScript 파일

### 간단한 예제를 만들어볼게요.

#### 1. `index.html` 파일:
```html
<!DOCTYPE html>
<html lang="en">
<head>
    <meta charset="UTF-8">
    <meta name="viewport" content="width=device-width, initial-scale=1.0">
    <title>Simple TypeScript Example</title>
</head>
<body>
    <div id="output"></div>
    <h1>Simple TypeScript Example</h1>

    <!-- 컴파일된 JS 파일을 여기서 로드 -->
    <script defer src="app.js"></script>
</body>
</html>
```

#### 2. `app.ts` 파일 (TypeScript 코드):
```typescript
// 두 숫자를 더하는 함수
function sum(a: number, b: number): number {
    return a + b;
}

const num1: number = 10;
const num2: number = 20;
const result: number = sum(num1, num2);

// 결과를 콘솔에 출력
console.log(`Sum: ${result}`);

// 결과를 웹 페이지에 출력
document.getElementById('output')!.innerText = `The sum of ${num1} and ${num2} is ${result}`;
```

#### 3. `app.js` 파일 (TypeScript에서 컴파일된 JavaScript):
TypeScript 파일을 컴파일하면 자동으로 생성됩니다. 하지만 예시로 작성된 JavaScript 코드는 다음과 같을 것입니다:
```javascript
function sum(a, b) {
    return a + b;
}

var num1 = 10;
var num2 = 20;
var result = sum(num1, num2);

console.log("Sum: " + result);

document.getElementById('output').innerText = "The sum of " + num1 + " and " + num2 + " is " + result;
```

### 동작 과정:
1. **`app.ts`** 파일을 작성한 후, `tsc` 명령어를 실행하여 TypeScript 파일을 컴파일합니다:
   ```bash
   tsc app.ts
   ```
   그러면 `app.js` 파일이 생성됩니다.

2. **`index.html`**에서 컴파일된 `app.js` 파일을 `<script defer src="app.js"></script>`로 로드하여 웹 페이지에서 사용합니다.

이제 웹 페이지를 브라우저에서 열면, HTML 요소에 계산 결과가 표시될 것입니다.
